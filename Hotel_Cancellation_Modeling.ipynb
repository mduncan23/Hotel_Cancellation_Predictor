# Data Preperation

```python
# pulling in data
df = pd.read_csv('./Data/Hotel Reservations.csv')
```

## Feature Engineering

### Checking Date Range - Adding Date Column

Checking the date range for the data. There is data for February 29th for a year that does not have a leap year. To account for this, I have adjusted the date to be February 28th for any days that were incorrectly labeled.

```python
# checking for any leap years
df['arrival_year'].min(), df['arrival_year'].max()
```

```python
# checking for any incorrect dates in February (no leap years in dataset)
df[df['arrival_month'] == 2].arrival_date.max()
```

```python
# replacing incorrectly labeled 2/29 with 2/28
df.loc[df['arrival_month'] == 2, 'arrival_date'] = df['arrival_date'].replace(29, 28)
```

```python
# adding a combined date column
df['date'] = pd.to_datetime(df[['arrival_year', 'arrival_month', 'arrival_date']].
                            rename(columns={'arrival_year': 'year',
                                            'arrival_month': 'month', 
                                            'arrival_date': 'day'}))
```

```python
# checking date range
df.date.min(), df.date.max()
```

### Adding new feature `total_stay`

Adding feature `total_stay` to account for all weekday and week end stays together.

```python
df['total_stay'] = df['no_of_week_nights'] + df['no_of_weekend_nights']
```

```python
df.head()
```

### Adding New Feature `total_stay_price`

Adding feature `total_stay_price` to account for the total price for all days stayed.

```python
df['total_stay_price'] = df['total_stay'] * df['avg_price_per_room']
```

```python
df.head()
```

## Changing dtype for use later on with scaling. 

I only want to scale continuous features and will convert them to float type to select just those features later on.

```python
# cols to convert to float type
cols_to_float = ['no_of_weekend_nights', 'no_of_week_nights', 'lead_time', 
                 'no_of_previous_cancellations', 'no_of_previous_bookings_not_canceled',
                 'avg_price_per_room', 'no_of_special_requests', 'total_stay',
                 'total_stay_price']
```

```python
# converting cols to float type
for col in cols_to_float:
    df[col] = df[col].astype(float)
```

```python
df.info()
```

## Cleaning up the Target Variable

```python
df['booking_status'].value_counts()
```

Lets recode these to numerical values.

- 0 for not_canceled
- 1 for canceled

```python
df['booking_status'] = df['booking_status'].map({'Not_Canceled':0, 'Canceled':1})
```

## Setting up a pipeline

```python
float_pipe = Pipeline([
    ('ss', StandardScaler())
])
```

```python
cat_pipe = Pipeline([
    ('ohe', OneHotEncoder(drop='first', sparse_output=False))
])
```

```python
ColTrans = ColumnTransformer(transformers=[
    ('float_pipe', float_pipe, selector(dtype_include=float)),
    ('cat_pipe', cat_pipe, selector(dtype_include=object))
], remainder='passthrough')
```

## Setting up Train/Test Split

```python
df.info()
```

```python
X = df.drop(['Booking_ID', 'booking_status', 'date'], axis=1)
y = df['booking_status']
```

```python
# creating initial validation set
X_train, X_val, y_train, y_val = train_test_split(X, 
                                                  y, 
                                                  test_size=.2, 
                                                  random_state=42)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, 
                                                    y_train,
                                                    test_size=.2,
                                                    random_state=42)
```

### Creating Score Cards to Keep Track of Results

```python
score_card = pd.DataFrame(columns=['Name', 'Accuracy', 'Precision', 'Recall', 
                                   'F1 Score', 'ROC/AUC Score'])

cv_score_card = pd.DataFrame(columns=['Name', 'Train Accuracy', 'Test Accuracy',
                                      'Train Precision', 'Test Precision',
                                      'Train Recall', 'Test Recall',
                                      'Train F1 Score', 'Test F1 Score',
                                      'Train ROC/AUC Score', 'Test ROC/AUC Score'])
```
